```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
df.head()
df
```

To check the distribution of age use seaborn - data visualization library

```python
import seaborn as sns
```

```python
sns.distplot(df['Age'], kde = True, hist = True) #distribution plot
```

```python
sns.distplot(df['Age'], bins = 30, color = "red") #more bins more precise
```

```python
sns.distplot(df['FastingBS'], kde = True, hist = True) # values are either 0 or 1
```

Pie chart for categorical values

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind = 'pie', autopct = '%.1f')
#autopct to show data text
```

```python
df.groupby('ChestPainType').size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby('HeartDisease').size().plot(kind = 'pie', autopct = '%.1f')
```

Violin plot when multiple columns need to be analysed for example how many females are there that have the heart disease and similarly for males and see if there is any correlation

```python
sns.violinplot(df['Age']) #middle white line is the mean or the average
```

```python
sns.violinplot(y = df['Sex'], x = df['HeartDisease'])
```

0 = no heart disease and 1 means heart disease so females are very less compared to males that have disease so males have more heart disease

```python
sns.violinplot(y = df['Cholesterol'], x = df['HeartDisease'], color = "orange")
```

Finding correlation - Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

JointPlot - Correlation between all the columns so total 7x7 = 49 graphs

```python
sns.jointplot(x = 'Age', y = "MaxHR", data = df)
```

```python
sns.jointplot(x = 'Age', y = "Age", data = df)
```

Completely correlated

```python
sns.jointplot(x = 'HeartDisease', y = "Oldpeak", data = df)
```

Zero correlation

```python
sns.jointplot(x = 'Age', y = "MaxHR", data = df, kind = "hex", color = "red")
```

middle area is where more of the data converges - darker the red higher the correlation

```python
sns.jointplot(x = 'Age', y = "MaxHR", data = df, kind = "reg") #best fit reg line
```

pair plot = correlation for a pair of two for all the combinations

```python
sns.pairplot(df)
```
